# Step 3: Handle Missing Values

In this step, the dataset is checked for missing (null) values. If any missing values are found, appropriate cleaning techniques such as filling or dropping them can be applied.

```python
# Check missing values in each column
missing_values = df.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)
```

```python
# Total missing values
total_missing = df.isnull().sum().sum()

print(f"Total Missing Values: {total_missing}")
```

```python
# Percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100

print("Percentage of Missing Values:")
print(missing_percentage)
```

```python
if total_missing > 0:
    df = df.dropna()
    print("Missing values removed.")
else:
    print("No missing values found. Dataset is already clean.")
```

```python
empty_rows = df.isnull().all(axis=1).sum()

print(f"Completely Empty Rows: {empty_rows}")
```

# Step 5: Remove Duplicate Records

Duplicate records can affect data analysis and lead to incorrect results. In this step, the dataset is checked for duplicate rows and duplicate records are removed if present.

```python
# Count duplicate rows
duplicate_count = df.duplicated().sum()

print(f"Number of Duplicate Records: {duplicate_count}")
```

```python
# Store dataset shape before removing duplicates
rows_before = df.shape[0]

print("Rows before removing duplicates:", rows_before)
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

print("Duplicate records removed successfully.")
```

```python
rows_after = df.shape[0]

print("Rows after removing duplicates:", rows_after)
print("Duplicates Remaining:", df.duplicated().sum())
```

```python
removed_rows = rows_before - rows_after

print(f"Total Duplicate Rows Removed: {removed_rows}")
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Create Price column
df["Price"] = df["Sales"] / df["Quantity"]

# Create total_amount column
df["total_amount"] = df["Price"] * df["Quantity"]

# Display the new columns
df[["Sales", "Quantity", "Price", "total_amount"]].head()
```
